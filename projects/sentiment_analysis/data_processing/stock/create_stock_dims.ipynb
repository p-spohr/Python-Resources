```python
top_50_df = pd.read_csv('top_50_news_volume.csv')
```

```python
for comp in top_50_df.itertuples():
    try:
        comp_fullname = yf.Ticker(comp.ticker).info['longName']
        print(f'{comp_fullname} ({comp.company}) has a valid ticker.')
    except Exception as e:
        print(f"{comp.company}: {comp.ticker} has error type: {type(e).__name__}")
```

```python
sent_df.loc[sent_df.company == 'stmicroelectronics'].head()
```

```python
target_metadata = ['symbol', 'shortName', 'longName', 'country', 'industry', 'sector', 'currency', 'financialCurrency', 'language', 'region', 'typeDisp', 'market', 'fullExchangeName', 'exchange']
```

```python
comp_dim_df = pd.DataFrame(columns= target_metadata)
for comp in top_50_df.itertuples():
    try:
        # get company info from yf
        comp_dim = yf.Ticker(comp.ticker).info
        result_df = pd.DataFrame([{key: comp_dim[key] for key in target_metadata if key in comp_dim}])
      
        # concatenate to company dim df
        comp_dim_df = pd.concat([comp_dim_df, result_df], ignore_index= True)
        
    except Exception as e:
        print(f"{comp.company}: {comp.ticker} has error type: {type(e).__name__}")
        print(e)
```

```python
comp_dim_df.to_csv('top_50_dim.csv', index= False)
```
