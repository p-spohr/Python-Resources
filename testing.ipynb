```python
class Animal():
    def __init__(self):
        pass
```

```python
animal_1 = Animal()
```

```python
def bark(animal):
    return "Woof!"
```

```python
bark(animal_1)
```

```python
def meow(animal):
    return "Meow!"
```

```python
meow(animal_1)
```

```python
class Dog(Animal):
    def __init__(self):
        self.bark_fn = bark
    def bark(self):
        return self.bark_fn(self)
```

```python
dog_1 = Dog()
```

```python
dog_1.bark()
```
