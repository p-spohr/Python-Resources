```python
import http.client
import urllib
```

```python
# local host use HTTP not HTTPS
conn = http.client.HTTPConnection('localhost', 8000)
conn.request('GET', '/hi.txt')

r1 = conn.getresponse()
print('status: ' + str(r1.status))
print('reason: ' + str(r1.reason))
print(r1.read().decode())
```

```python
# local host use HTTP not HTTPS
conn = http.client.HTTPConnection('localhost', 8000)
conn.request('PUT', '/', 'Hello local server!')

r1 = conn.getresponse()
print('status: ' + str(r1.status))
print('reason: ' + str(r1.reason))
print(r1.read().decode())
```
